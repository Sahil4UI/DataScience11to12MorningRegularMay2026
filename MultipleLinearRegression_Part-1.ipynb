```python
def cost_function(X,y,B):
    return (np.sum((np.dot(X,B) - y) ** 2)) / (2 * n)
```

```python
cost_function(X,y,B)
```

```python
def gradient_descent(X,y,B,epochs,alpha):
    cost_history = [0] * epochs
    for i in range(epochs):
        y_pred = X.dot(B)
        loss = y_pred - y
        gradient = X.T.dot(loss) / n
        B = B - alpha * gradient
        cost_history[i] = cost_function(X,y,B)
        print("ch =>",cost_history[i])
        
    return cost_history, B
```

```python
cost,b = gradient_descent(X,y.flatten(),B,epochs=10000,alpha=0.0001)
```
