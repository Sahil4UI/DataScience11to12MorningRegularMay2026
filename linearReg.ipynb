```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/SaratogaHouses.csv")
```

```python
x = data["livingArea"]
y = data["price"]
```

```python
x_ = np.mean(x)
y_ = np.mean(y)
```

```python
num=0
den=0
for i in range(len(x)):
    num += (x[i]-x_)*(y[i]-y_)
    den += (x[i]-x_)**2

b1 = num/den
    
```

```python
b0 = y_-b1*x_
```

```python
x_new = x[0]
y_pred = b0+b1*x_new
print(y_pred)
print("Error : ",y[0]-y_pred)
```
